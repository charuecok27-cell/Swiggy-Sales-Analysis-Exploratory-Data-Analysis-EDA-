### BUILDING KPI'S

##### TOTAL SALES

```python
total_sales = df["Price (INR)"].sum()
print("Total Sales (INR):",  round(total_sales,2))
```

This represents the total revenue generated from all orders in the dataset. A total sales value of over ₹5.3 crore indicates a high transaction volume and strong platform activity, suggesting that Swiggy has significant market penetration and consistent customer demand across cities and restaurants.

#### AVERAGE RATING

```python
average_rating= df["Rating"].mean()
print("Average Rating:", round(average_rating,1))
```

An average rating of 4.3 reflects high customer satisfaction. Most customers are likely happy with food quality, delivery experience, and service

#### AVERAGE ORDER VALUE

```python
avg_order_value = df["Price (INR)"].mean()
print("Avg Order Value(INR):", round(avg_order_value,2))
```

On average, customers spend around ₹268 per order, indicating that most orders fall in the low-to-mid price range.

#### RATINGS COUNT

```python
ratings_count= df["Rating Count"].sum()
print("Ratings Count:", round(ratings_count,2))
```

Over 5.5 million ratings show extremely high customer engagement and feedback participation.

#### TOTAL ORDERS

```python
total_orders = df["Order Date"].count()
print("Total Orders:",total_orders )
```

Nearly 2 lakh orders were placed in the dataset period, showing strong and consistent platform usage.

Overall, the KPIs together paint a picture of a high-volume, mass-market, and customer-satisfied platform. Swiggy is generating over ₹5.3 crore in revenue from nearly 2 lakh orders, with customers spending around ₹268 per order and rating their experience highly (4.3/5). The presence of 5.5 million ratings highlights strong customer engagement and trust. Collectively, this indicates a scalable and mature business model, where growth opportunities lie in increasing average order value, leveraging customer feedback for quality improvements, and upselling premium offerings to boost revenue without needing proportional increases in user base.

### CHARTS

#### MONTHLY SALES TREND

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["YearMonth"]= df["Order Date"].dt.to_period("M").astype(str)
monthly_revenue = df.groupby("YearMonth")["Price (INR)"].sum().reset_index()

plt.figure()
plt.plot(monthly_revenue["YearMonth"], monthly_revenue["Price (INR)"])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Total Revenue (INR)")
plt.title("Monthly Revenue Trend")
plt.tight_layout()
plt.show()
```

The monthly revenue shows moderate fluctuations but an overall stable pattern. Revenue starts high in January, drops noticeably in February( dip may indicate post-holiday slowdown or fewer promotions), then gradually recovers through March and April, peaks again around May, dips slightly in June, and rises again towards August. This suggests that while demand is consistent, there are seasonal or event-based effects influencing customer spending.

### DAILY SALES TREND

```python
df["DayName"] = pd.to_datetime(df["Order Date"]).dt.day_name()
daily_revenue = (df.groupby("DayName")["Price (INR)"].sum().reindex(["Monday","Tuesday","Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]))
```

```python
plt.figure(figsize=(10,5))
plt.bar(daily_revenue.index, daily_revenue.values, color='pink')
plt.title("Daily Revenue Trend (Mon-Sun)")
plt.xlabel("Day")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=30)

plt.show()
```

Revenue is fairly evenly distributed across all days of the week, with a slight increase towards Thursday, Saturday, and Sunday. Saturday records the highest revenue, while Tuesday is marginally lower than other days.

Weekends generate slightly more revenue, likely due to people ordering more when they are at home.
However, the small differences show that Swiggy experiences strong daily demand throughout the week, not just on weekends.

### TOTAL SALES BY FOOD TYPE (VEG/NON-VEG)

```python
non_veg_keyword= ["chicken", "mutton", "egg", "fish", "prawn", "meat", "biryani", "kabab", "kebab", "non-veg", "non veg"]

df["Food Category"]= np.where(df["Dish Name"].str.lower().str.contains('|'.join(non_veg_keyword),na=False),"Non-Veg","Veg")
```

```python
food_revenue= (df.groupby("Food Category")["Price (INR)"].sum().reset_index())
```

```python
fig = px.pie(food_revenue, values= "Price (INR)", names="Food Category", hole= 0.5, title="Revenue Contribution: Veg vs Non-Veg")
fig.update_traces(textinfo="percent+label", pull=[0.05,0])

fig.update_layout(height=500, margin= dict(t=60, b=40, l=40, r=40))
fig.show()
```

The chart shows that Vegetarian food contributes 62.7% of total revenue, while Non-Vegetarian contributes 37.3%. This indicates that a significant majority of Swiggy’s revenue comes from vegetarian dishes.

### TOTAL SALES BY STATE

```python
fig =px.bar(df.groupby("State", as_index=False)["Price (INR)"].sum().sort_values("Price (INR)", ascending=False), x="Price (INR)", y="State", orientation="h", title="Revenue by State(INR)")
fig.update_layout(height=600, yaxis= dict(autorange="reversed"))
fig.show()
```

Karnataka is the highest revenue-generating state by a large margin.
This indicates extremely strong adoption of Swiggy in Karnataka, most likely driven by cities like Bangalore, which has a high concentration of working professionals, tech employees, and frequent online food consumers.

Next top contributors are Uttar Pradesh, Telangana, and Maharashtra.
These are large, densely populated states with major urban centers such as Hyderabad, Mumbai, Pune, Noida, and Lucknow, leading to high order volumes.

Delhi, Gujarat, and Punjab form a strong second tier.
These states show solid revenue, reflecting stable and consistent demand.

Mid-level states include West Bengal, Tamil Nadu, Rajasthan, and Madhya Pradesh.
These markets are performing reasonably well but still have growth potential, especially outside capital cities.

North-Eastern and smaller states (Sikkim, Nagaland, Mizoram, Tripura) contribute the least.
This likely reflects:

Lower population density

Limited restaurant availability

### QUARTERLY SUMMARY

```python
df["Order_Date"]= pd.to_datetime(df["Order Date"])
df["Quarter"] = df["Order_Date"].dt.to_period("Q").astype(str)
quarterly_summary= (df.groupby("Quarter", as_index=False).agg(Total_Sales= ("Price (INR)", "sum"), Average_Rating= ("Rating", "mean"), Total_Orders= ("Order_Date", "count")).sort_values("Quarter"))

quarterly_summary["Total_Sales"]= quarterly_summary["Total_Sales"].round(0)
quarterly_summary["Average_Rating"]= quarterly_summary["Average_Rating"].round(2)

quarterly_summary
```

Q1 shows strong performance with high revenue and order volume. This indicates a solid start to the year, likely driven by consistent customer demand and high service quality.

Q2 is the best-performing quarter, showing growth in both revenue and orders compared to Q1. This suggests successful customer retention and possibly effective promotions or seasonal demand.

Q3 shows a significant drop in both sales and orders. However, customer satisfaction remains unchanged, indicating that the decline is likely due to lower demand or incomplete data for the quarter, rather than service issues.

# CONCLUSION

The overall analysis presents Swiggy as a high-volume, stable, and metro-driven platform with strong customer satisfaction and scalable growth potential. With total sales exceeding ₹5.3 crore from nearly 2 lakh orders, the platform operates at significant scale, supported by a consistently high average customer rating of 4.3, indicating strong service quality and customer trust.

Revenue patterns show remarkable stability across time. Monthly and daily trends reflect only mild fluctuations, suggesting that food delivery has become a habitual, necessity-based service rather than a seasonal luxury. The strongest activity occurs in the first half of the year (Q1–Q2), while the dip in Q3 appears to be demand-related rather than quality-driven, as customer satisfaction remains unchanged.

From a product perspective, vegetarian food dominates the platform, contributing nearly 63% of total revenue, which aligns with Indian consumption patterns and highlights Swiggy’s dependence on mass, affordable daily meals. Non-vegetarian food, while smaller in share, represents a high-value growth segment, particularly for premium pricing and weekend demand.

Geographically, revenue is highly concentrated in urban regions. Karnataka alone leads state-wise sales, driven primarily by Bengaluru, which emerges as the single largest revenue-generating city. However, the strong performance of cities like Lucknow signals a major opportunity in Tier-2 markets, where digital adoption is accelerating and competition is relatively lower.

In summary, Swiggy’s business model is currently powered by metro consumption, daily low-to-mid value orders, and a predominantly vegetarian customer base, all sustained by strong customer satisfaction. The platform’s future growth lies not in expanding the user base aggressively, but in increasing order value, monetizing non-veg and premium segments, and scaling deeper into Tier-2 and Tier-3 cities.
